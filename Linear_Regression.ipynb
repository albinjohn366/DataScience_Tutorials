```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
%matplotlib inline
```

```python
from sklearn.datasets import load_boston

boston = load_boston()
```

```python
print(boston.DESCR)
```

Here we will consider MEDV as the target or label.

Lett us have a visualization of the dataset

```python
sns.histplot(x=boston.target, color='brown')
plt.xlabel('Price')
plt.ylabel('Number of houses')
plt.show()
```

```python
# Before doing anything lets convert this scikiit object to a dataframe
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df['target'] = boston.target
```

Now let us visualiize the relationship between each of the features.

```python
sns.pairplot(boston_df)
```

Only RM and LSTAT has somewhat linear relationship with target.

RM is average number of rooms per dwelling. Let us see the relation between this data and target.

```python
sns.scatterplot(data=boston_df, x='RM', y='target', color='orange')
plt.xlabel('Number of  rooms per dwelling')
plt.ylabel('Price')
plt.show()
```

Both of these have a linear relationship

LSTAT is percentage of lower status of people.

```python
sns.scatterplot(data=boston_df, x='LSTAT', y='target', color='orange')
plt.xlabel('Percentage of lower status people')
plt.ylabel('Price')
plt.show()
```

When the percentage of lower status people go up, the price go down.
